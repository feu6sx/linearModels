**Q1.** Please answer the following questions in your own words.

1. What makes a model "linear"? "Linear" in what?

A model is linear if the relationship between the input features and the output is a straight line, meaning the model’s parameters combine additively without any complex functions like exponents or multiplication between parameters. It's "linear" in the parameters, not necessarily the features.

2. How do you interpret the coefficient for a dummy/one-hot-encoded variable? (This is a trick question, and the trick involves how you handle the intercept of the model.)

When you have a dummy variable like one-hot encoding, the coefficient tells you how much that category’s outcome is different from the reference category (the one you left out when creating dummies). The intercept is actually the predicted value for that reference category. So, if the dummy variable’s coefficient is 5 it means the category represented by that dummy is 5 units higher than the reference category’s predicted value. It’s all about comparing the new categories to the reference one.
   
3. Can linear regression be used for classification? Explain why, or why not.

You can technically use linear regression for classification, but it’s not great for the job. It predicts continuous values, so for something like binary classification, it might give you numbers like 1.5 or -0.2, which don't really make sense as class labels (you want 0 or 1). Plus, it doesn't handle things like decision boundaries well, which is important for classification. That’s why models like logistic regression or decision trees are usually the go-to for classification problems.

4. What are signs that your linear model is over-fitting?

Overfitting happens when your model does really well on the training data but poorly on the test data, meaning it's memorizing the training set rather than learning general patterns. It can also happen if you're using too many features, especially ones that don’t add value. Another sign is if the training error is much lower than the validation error. To fix it, try regularization, cross-validation, or simplifying the model.

5. Clearly explain multi-colinearity using the two-stage least squares technique.

Multicollinearity happens when your predictors are super correlated with each other, making it hard to figure out their individual effects. The Two-Stage Least Squares (2SLS) method helps fix this. First, you find variables that are related to your predictors but not the outcome to replace the correlated predictors. Then, in the second stage, you use the predicted values from the first stage to run the regression. This way, you reduce the correlation issue and get more reliable results.

6. How can you incorporate nonlinear relationships between your target/response/dependent/outcome variable $y$ and your features/control/response/independent variables $x$ into your analysis?

To capture nonlinear relationships between your target variable and features, you’ve got a few options. You can use polynomial features to add curves to your model or add interaction terms to capture how features work together. If the relationship looks more like a log or exponential curve, try applying logarithmic or exponential transformations to your features. If your data is bounded between 0 and 1, you might want to try an arcsine transformation to help linearize the relationship by stretching out the lower values and compressing the higher ones.

7. What is the interpretation of the intercept? A slope coefficient for a variable? The coefficient for a dummy/one-hot-encoded variable?

The intercept is the predicted value of y when all your features are zero. The slope coefficient tells you how much y changes for a one-unit increase in a
feature. For example, if the slope is 2, y goes up by 2 for each unit increase in x. For a dummy variable, the coefficient shows how much y changes between that category and the reference category. If it’s 5, it means that category’s predicted y is 5 units higher than the reference category.

```python
import pandas as pd

df_heart=pd.read_csv('data/heart_hw.csv')
df_heart_control=df_heart[df_heart['transplant'] == 'control']
control_survive=sum(df_heart_control['y'])/len(df_heart_control)
print(f'The proportion of control-group survivors is {control_survive:.4f}.')

df_heart_treat=df_heart[df_heart['transplant'] == 'treatment']
treat_survive=sum(df_heart_treat['y'])/len(df_heart_treat)
print(f'The average treatment effect is equal to {treat_survive-control_survive}.')
```

```python
import numpy as np

def slr(x,y): 
    """ Single linear regression function. """
    x_bar = np.mean(x)
    y_bar = np.mean(y)
    b1 = np.inner(x-x_bar,y-y_bar)/np.inner(x-x_bar,x)
    b0 = y_bar - b1*x_bar
    y_hat = b0 + b1*x
    residuals = y - y_hat
    return({'b0':b0,'b1':b1,'y_hat':y_hat,'residuals':residuals})

df_heart['transplant_dummy'] = (df_heart['transplant'] == 'treatment').astype(int)
x = df_heart['transplant_dummy']
y = df_heart['y']

regression_results = slr(x, y)

print(f"Intercept (Control Group Survival Rate): {regression_results['b0']:.4f}")
print(f"Coefficient on Transplant (ATE): {regression_results['b1']:.4f}")

# The constant/intercept are indentical 
# Both pieces of code are estimating the average treatment effect (ATE) of a transplant. The first block just calculates the proportion 
# of survivors in each group (control and treatment) and takes the difference. It’s basically doing a simple comparison of means. The 
# second block, on the other hand, sets up a linear regression with a dummy variable for treatment (1 for treatment, 0 for control), 
# and the coefficient on that dummy variable gives you the ATE. It’s the same idea, but using OLS regression to model the difference between
# groups instead of just calculating proportions directly. Both give you the same result in the end.
```

```python
def slr_with_age(x, y, age):
    """ Single linear regression with age as a covariate. """
    X = np.vstack([np.ones(len(x)), x, age]).T  # add intercept, transplant dummy, and age
    b = np.linalg.inv(X.T @ X) @ X.T @ y  # OLS estimation
    y_hat = X @ b  # predicted values
    residuals = y - y_hat
    return {'b0': b[0], 'b1': b[1], 'b2': b[2], 'y_hat': y_hat, 'residuals': residuals}

x = df_heart['transplant_dummy']
y = df_heart['y']
age = df_heart['age']  

regression_results = slr_with_age(x, y, age)

print(f"Intercept (Control Group Survival Rate, adjusted for age): {regression_results['b0']:.4f}")
print(f"Coefficient on Transplant (ATE, adjusted for age): {regression_results['b1']:.4f}")
print(f"Coefficient on Age: {regression_results['b2']:.4f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# features and target variable
X = df_heart[['transplant_dummy', 'age']]
y = df_heart['y']

# polynomial features for age 
poly = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly.fit_transform(X[['age']])  
transplant_dummy = X['transplant_dummy'].values.reshape(-1, 1)

# transplant dummy and interaction term
X_poly = np.column_stack([X_poly, transplant_dummy])  # Add transplant dummy
X_poly = np.column_stack([X_poly, X_poly[:, 0] * transplant_dummy[:, 0]])

# split data into training and testing (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# fit w linear regression
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error on the test set: {mse:.4f}")

# prediction grid
age_range = np.linspace(df_heart['age'].min(), df_heart['age'].max(), 100)
age_poly = poly.transform(age_range.reshape(-1, 1))  

control_pred = np.column_stack([age_poly, np.zeros((100, 1)), age_poly[:, 0] * 0])  # control: transplant_dummy = 0
treatment_pred = np.column_stack([age_poly, np.ones((100, 1)), age_poly[:, 0] * 1])  # treatment: transplant_dummy = 1
control_survival = model.predict(control_pred)
treatment_survival = model.predict(treatment_pred)

plt.figure(figsize=(10, 6))
plt.plot(age_range, control_survival, label="Control Group", color='blue')
plt.plot(age_range, treatment_survival, label="Treatment Group", color='red')
plt.title("Predicted Survival Probability by Age (Control vs Treatment)")
plt.xlabel("Age")
plt.ylabel("Predicted Survival Probability")
plt.legend()
plt.grid(True)
plt.show()

# Using models to decide who gets an organ transplant raises a lot of ethical and practical concerns. One big issue is bias; if the model is trained
# on historical healthcare data, it might reflect existing inequalities, unintentionally prioritizing certain groups over others. Transparency is 
# another concern. If a patient is denied a transplant, they should have a clear explanation, not just “the algorithm decided.” There’s also the 
# question of data quality; if the model relies on incomplete or outdated data, it could make unfair decisions. For example, someone who had a 
# temporary health setback might be unfairly ranked lower. And finally, accountability matters. If an algorithm makes a mistake, who takes 
# responsibility? These models can be helpful for guiding decisions, but they shouldn’t be the only factor in something as critical as organ allocation.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# for singular regression, just use 'Study_Hours_per_Week' as the feature
X_singular = df_student[['Study_Hours_per_Week']]
y = df_student['Final_Exam_Score']

# for multiple regression, use several features
X_multiple = df_student[['Study_Hours_per_Week', 'Attendance_Rate',
                         'Past_Exam_Scores', 'Parental_Education_Level',
                         'Internet_Access_at_Home', 'Extracurricular_Activities']]

# One-Hot Encoding for categorical variables
categorical_cols = ['Parental_Education_Level', 'Internet_Access_at_Home', 'Extracurricular_Activities']
preprocessor_multiple = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_cols)
    ],
    remainder='passthrough'  
)

preprocessor_singular = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), [])  # no categorical columns here
    ],
    remainder='passthrough'  # pass through the numeric 'Study_Hours_per_Week'
)

# split data into train and test sets (80/20)
X_train_singular, X_test_singular, y_train, y_test = train_test_split(X_singular, y, test_size=0.2, random_state=42)
X_train_multiple, X_test_multiple, _, _ = train_test_split(X_multiple, y, test_size=0.2, random_state=42)

# simple linear regression (singular regression) model
lr_singular = Pipeline(steps=[('preprocessor', preprocessor_singular),
                              ('regressor', LinearRegression())])
lr_singular.fit(X_train_singular, y_train)
lr_singular_preds = lr_singular.predict(X_test_singular)
lr_singular_sse = np.sum((y_test - lr_singular_preds) ** 2)

# multiple linear regression model
lr_multiple = Pipeline(steps=[('preprocessor', preprocessor_multiple),
                               ('regressor', LinearRegression())])
lr_multiple.fit(X_train_multiple, y_train)
lr_multiple_preds = lr_multiple.predict(X_test_multiple)
lr_multiple_sse = np.sum((y_test - lr_multiple_preds) ** 2)

# model comparison
results = {
    'Simple Linear Regression (Singular)': lr_singular_sse,
    'Multiple Linear Regression': lr_multiple_sse
}

best_model = min(results, key=results.get)
best_sse = results[best_model]

best_model, best_sse

# Multiple regression works better because it considers more factors that could affect the outcome. When you only use one variable, 
# like "Study_Hours_per_Week," you're limiting the model to just that single factor. But with multiple regression, you're using other 
# relevant features like "Attendance_Rate," "Past_Exam_Scores," and "Parental_Education_Level," which give the model more information 
# to make a better prediction.

# I learned that predicting something complex, like exam scores, isn’t as simple as just looking at one factor. It's kind of like how 
# we can't just rely on one study method for success in college. We need to consider multiple things—like attendance, past performance,
# and even things outside the classroom, like parental support or extracurricular activities. It really clicked for me that multiple 
# regression is like getting the full picture instead of trying to predict the outcome with only half the information. 
# More data = better predictions.
```

We derived the OLS estimator for single linear regression by minimizing
$$
SSE(b_0, b_1) = \frac{1}{N}\sum_{i=1}^N (y_i - b_0 - b_1 x_i)^2
$$
with solution
$$
\hat{b}_0 = \bar{y} - \hat{b}_1 \bar{x}, \quad \hat{b}_1 = \dfrac{\frac{1}{N} \sum_{i=1}^N (x_i-\bar{x})(y_i - \bar{y})}{s_x^2}.
$$

When you step back and think about it, this is a bit weird: The algorithm is computing sample means, variances, and covariances, and using those to create a predictive model. The data themselves arguably vanish from the solution. This is elegant, this is strange.

Instead, let $k(z)$ be a kernel function, such as the Gaussian
$$
k(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}
$$
or uniform
$$
k(z) = \begin{cases}
1/2, & |z| \le \frac{1}{2}\\
0, & \text{otherwise.}
\end{cases}
$$
We'll instead minimize, or each predictor value $x$,
$$
SSE(\hat{y}(x)) = \dfrac{1}{N} \sum_{i=1}^N \left\lbrace y_i - \hat{y}(x)\right\rbrace^2 \dfrac{1}{h} k \left( \dfrac{x-x_i}{h} \right).
$$

1. Show that the optimal predictor is
$$
\hat{y}(x) = \dfrac{ \frac{1}{Nh} \sum_{i=1}^N  y_i k \left( \dfrac{x-x_i}{h} \right) }{\frac{1}{Nh} \sum_{i=1}^N k \left(  \dfrac{x-x_i}{h} \right)}
$$
This has many names, but let's call it the local constant least squares (LCLS) estimator, as opposed to ordinary least squares (OLS).
